```python
old_text1=soup.find_all('h3')
print(old_text1)
```

```python
#replacing all Data Scientist as AWS Engineer And all other text as Associate Engineer
newrecords=[]
for record in old_text1:
  if record.string:
    new_text1=record.string.replace_with(record.string.replace('Data Scientist', 'AWS Engineer'))
   #new_text1=record.replace_with("AWS Engineer")
    newrecords.append(new_text1)
  else:
   #new_text1=record.string.replace_with(record.string.replace('', 'AWS Engineer')) 
   new_text1=record.string.replace_with("Associate Engineer")
   newrecords.append(new_text1)

print(newrecords)
```
